We can also calculate the **first derivative** of the curve, or the **infection rate**, for any country and plot it 


The derivative  $f'(a)$ of the function $f(x)$ at the point $x=a$ is the limit

$$f'(a) = \lim{x \to a} \frac{f(x) - f(a)}{x - a}$$


Here we will do this with the two countries presented above: `Brazil` and the `United States`. We can set variables named `country_infection_rate` for each country similar to how we made the time series above.

We can find the maximum infection rate for all the countries by implementing a **for loop** and **appending** our original data frame 

First, we can set a variable, `countries`, as a list of all the countries represented in the dataframe. The next step is to create an open array for the maximum infection rates we will calculate. To do that, all that is needed is a new variable, `max_infection_rates`, and open brackets `[]`. 


The **for loop** provided below will then be used to **append**, or add, a maximum infection rate column of data to the dataframe we have been working with, `corona_dataset_aggregated_sum`.

```python
countries = list(corona_dataset_aggregated_sum.index)

max_infection_rates = []
for country in countries:
    max_infection_rates.append(corona_dataset_aggregated_sum.loc[country].diff().max())

corona_dataset_aggregated_sum['max infection rate'] = max_infection_rates
```
